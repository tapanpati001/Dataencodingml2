Nominal/OHE Encoding

Nominal encoding is a technique used to transform categorical variables that have no intrinsic ordering into numerical values that can be used in machine learning models.
One common method for nominal encoding is one_hot encoding,which creates a binary vector for each category in the variable .

```python
import pandas as pd 
from sklearn.preprocessing import OneHotEncoder
```

```python
df=pd.DataFrame({'color':['red','blue','green','green','red','blue']})
```

```python
df
```

```python
#create an instance of one hot encoder
encoder=OneHotEncoder()
```

```python
encoder.fit_transform(df[['color']])
```

```python
encoder.fit_transform(df[['color']]).toarray()
```

```python
#fit the   encoder to the dataframe and transform the categorical variable 
encoded=encoder.fit_transform(df[['color']])
```

```python
import pandas as pd 
encoded_df=pd.DataFrame(encoded.toarray(),columns=encoder.get_feature_names_out())
```

```python
encoded_df
```

```python
pd.concat([df,encoded_df],axis=1)
```

Target Guided Ordinal Encoding 

It is a technique used to encode categorical varables based on their relationship with their target variables.
IN Target Guided Ordinal Encoding we replace each categorical variable based on their mean or median of the target variable for that category.

```python
import pandas as pd 
#create a simple dataframe with a categorical variable and a target variable
df=pd.DataFrame({'city':['New York','London','paris','Tokyo','New York','paris'],
                'price':[200,300,400,500,600,700]})
```

```python
df
```

```python
#Here price is the target variable and city is the categorical variable
```

```python
##calculate the mean price for ecach city
mean_price=df.groupby('city')['price'].mean().to_dict()#for converting into dictionary
mean_price
```

```python
##If there is outliers then we use the median value 
```

```python
##replace each city with their mean price
```

```python
df['city_encoded']=df['city'].map(mean_price)
```

```python
df
```

```python
import seaborn as sns 
df=sns.load_dataset('tips')
```

```python
df.head()
```

```python
mean_price=df.groupby('day')['total_bill'].mean().to_dict()
mean_price
```

```python
df['encoded_day']=df['day'].map(mean_price)
```

```python
df
```

```python
mean_price_time=df.groupby('time')['tip'].mean().to_dict()
```

```python
mean_price_time
```

```python
#replace ech time with their mean tip price
df['encoded_time']=df['time'].map(mean_price_time)
```

```python
df
```

```python
mean_price_size=df.groupby('time')['size'].mean().to_dict()
```

```python
mean_price_size
```

```python
df['encoded_time_size']=df['time'].map(mean_price_size)
```
